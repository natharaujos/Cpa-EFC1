##  Insertion Sort


```python

def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(0, 1000, (n_elementos))
  return A


def tempo(algoritmo, arranjo):
  inicio = time.time()
  comparacoes = algoritmo(arranjo)
  fim = time.time()
  tempoExecucao = fim - inicio

  return tempoExecucao, comparacoes

def insertionSort(s):
    compara = 0
    for i in range(1, len(s)):
        val = s[i]
        j = i - 1
        compara += 1
        while (j >= 0) and (s[j] > val):
            s[j+1] = s[j]
            j = j - 1
            compara += 1
        s[j+1] = val
    
    return compara

def main():
    auxAleatorio = []
    auxOrdenado = []
    auxInverso = []
    auxAleatorioComparacoes = []
    auxOrdenadoComparacoes = []
    auxInversoComparacoes = []
    tempoAleatorio = 0
    tempoInverso = 0
    tempoOrdenado = 0
    for a in range(5,1000):
      aleatorio = gera_arranjos_aleatorio(a)
      aux = tempo(insertionSort, aleatorio)
      auxAleatorio.append(aux[0])
      auxAleatorioComparacoes.append(aux[1])
    
    for a in range(5,1000):
      ordenado = np.arange(a)
      aux = tempo(insertionSort, ordenado)
      auxOrdenado.append(aux[0])
      auxOrdenadoComparacoes.append(aux[1])
    
    for a in range(5,1000):
      inverso = np.arange(a, 0, -1)
      aux = tempo(insertionSort, inverso)
      auxInverso.append(aux[0])
      auxInversoComparacoes.append(aux[1])

    print("")
    print("Gráfico de Tempo")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxInverso, label="Pior Caso")
    ax.legend()
    ax.plot(auxAleatorio, label="Aleatorio")
    ax.legend()
    ax.plot(auxOrdenado, label="Melhor Caso")
    ax.legend()
    plt.show()

    print("")
    print("Gráfico de Comparações")
    print("")
    fig, ax2 = plt.subplots()
    ax2.plot(auxInversoComparacoes, label="Pior Caso")
    ax2.legend()
    ax2.plot(auxAleatorioComparacoes, label="Aleatorio")
    ax2.legend()
    ax2.plot(auxOrdenadoComparacoes, label="Melhor Caso")
    ax2.legend()
    plt.show()

    print("")
    print("1 - c)")
    print("O comportamento assintótico do melhor caso no insertion sort é O(n)")
    print("O comportamento assintótico do pior caso no insertion sort é O(n^2)")

main()
```

##  Merge Sort


```python
def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(5, 1000, (n_elementos))
  return A

def tempo(algoritmo, arranjo):
  inicio = time.time()
  comparacoes = algoritmo(arranjo)[1]
  fim = time.time()
  tempoExecucao = fim - inicio

  return tempoExecucao, comparacoes

def mergeSort(arr):
    compara = 0
    compara += 1
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left, right = mergeSort(arr[:mid]), mergeSort(arr[mid:])
    compara += merge(left, right, arr.copy())

    return merge(left, right, arr.copy()), compara


def merge(left, right, merged):

    left_cursor, right_cursor = 0, 0
    compara = 0;
    while left_cursor < len(left) and right_cursor < len(right):
        compara += 1
        if left[left_cursor] <= right[right_cursor]:
            merged[left_cursor+right_cursor]=left[left_cursor]
            left_cursor += 1
        else:
            merged[left_cursor + right_cursor] = right[right_cursor]
            right_cursor += 1
    compara += 1
            
    for left_cursor in range(left_cursor, len(left)):
        merged[left_cursor + right_cursor] = left[left_cursor]
        compara += 1
    compara += 1
        
    for right_cursor in range(right_cursor, len(right)):
        merged[left_cursor + right_cursor] = right[right_cursor]
        compara += 1
    compara += 1

    return compara

def main():
    arrayAleatorio = gera_arranjos_aleatorio(100)
    arrayOrdenado = np.arange(100)
    arrayInverso = np.arange(99, 0, -1)
    resultadoAleatorio = mergeSort(arrayAleatorio)
    resultadoMelhorCaso = mergeSort(arrayOrdenado)
    resultadoPiorCaso = mergeSort(arrayInverso)
    arrayTempoAleatorioComp = resultadoAleatorio[0]
    arrayMelhorCasoComp = resultadoMelhorCaso[0]
    arrayPiorCasoComp = resultadoPiorCaso[0]
    auxAleatorio = []
    auxOrdenado = []
    auxInverso = []
    auxAleatorioComparacoes = []
    auxOrdenadoComparacoes = 0
    auxInversoComparacoes = 0
    tempoAleatorio = 0
    tempoOrdenado = 0
    tempoInverso = 0

    for a in range(5,1000):
      aleatorio = gera_arranjos_aleatorio(1000)
      auxAleatorio.append(tempo(mergeSort, aleatorio)[0])
      auxAleatorioComparacoes.append(tempo(mergeSort, aleatorio)[1])
      tempoAleatorio += tempo(mergeSort, aleatorio)[0]

    ordenado = np.arange(1000)
    auxOrdenado.append(tempo(mergeSort, ordenado)[0])
    auxOrdenadoComparacoes += tempo(mergeSort, ordenado)[1]
    tempoOrdenado += tempo(mergeSort, aleatorio)[0]
    inverso = np.arange(1000, 5, -1)
    auxInverso.append(tempo(mergeSort, inverso)[0])
    auxInversoComparacoes += tempo(mergeSort, inverso)[1]
    tempoInverso += tempo(mergeSort, aleatorio)[0]


    print("")
    print("O tempo no melhor caso foi de {:.5f}" .format(tempoInverso))
    print("")
    print("O tempo no pior caso foi de {:.5f}" .format(tempoOrdenado))
    print("")
    
    print("")
    print("O numero de comparações no melhor aleatório foi de", auxInversoComparacoes)
    print("")
    print("O numero de comparações no pior caso foi de", auxOrdenadoComparacoes )
    print("")

    print("")   
    print("Gráfico de Tempo")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxAleatorio, label="Caso aleatório")
    ax.legend()
    plt.show()
    print("")
    
    print("")   
    print("Gráfico de Comparações")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxAleatorioComparacoes, label="Caso aleatório")
    plt.show()
    ax.legend()


    print("")
    print("1 - c)")
    print("O comportamento assintótico do melhor caso no merge sort é O(n lg n)")
    print("O comportamento assintótico do pior caso no merge sort é O(n lg n)")
main()
```

##  Selection Sort


```python
def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(0, 100, (n_elementos))
  return A

def tempo(algoritmo, arranjo):
  inicio = time.time()
  comparacoes = algoritmo(arranjo)
  fim = time.time()
  tempoExecucao = fim - inicio

  return tempoExecucao, comparacoes

def selectionSort(s):
    compara = 0
    for i in range(0,len(s)-1):
        p=0
        mini=s[-1]
        compara += 1
        for j in range(i,len(s)):
            compara += 1
            if s[j]<=mini:
                mini=s[j]
                p=j
        s[i],s[p]=s[p],s[i]
    
    return compara

def main():
    arrayAleatorio = gera_arranjos_aleatorio(100)
    melhorCaso = np.arange(100)
    piorCaso = np.arange(99, 0, -1)
    resultadoAleatorio = selectionSort(arrayAleatorio)
    resultadoMelhorCaso = selectionSort(melhorCaso)
    resultadoPiorCaso = selectionSort(piorCaso)

    arrayTempoAleatorio = []
    arrayMelhorCaso = []
    arrayPiorCaso = []

    auxAleatorio = []
    auxOrdenado = []
    auxInverso = []
    auxAleatorioComparacoes = []
    auxOrdenadoComparacoes = []
    auxInversoComparacoes = []
    tempoAleatorio = 0
    tempoInverso = 0
    tempoOrdenado = 0
    for a in range(5,1000):
      aleatorio = gera_arranjos_aleatorio(a)
      auxAleatorio.append(tempo(selectionSort, aleatorio)[0])
      auxAleatorioComparacoes.append(tempo(selectionSort, aleatorio)[1])
      tempoAleatorio += tempo(selectionSort, aleatorio)[0]
      ordenado = np.arange(a)
      auxOrdenado.append(tempo(selectionSort, ordenado)[0])
      auxOrdenadoComparacoes.append(tempo(selectionSort, ordenado)[1])
      tempoOrdenado += tempo(selectionSort, ordenado)[0]
      inverso = np.arange(a, 0, -1)
      auxInverso.append(tempo(selectionSort, inverso)[0])
      auxInversoComparacoes.append(tempo(selectionSort, inverso)[1])
      tempoInverso += tempo(selectionSort, inverso)[0]

    print("")
    print("Gráfico de Tempo")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxOrdenado, label="Melhor Caso")
    ax.legend()
    ax.plot(auxAleatorio, label="Aleatorio")
    ax.legend()
    ax.plot(auxInverso, label="Pior Caso")
    ax.legend()
    plt.show()

    print("")
    print("Gráfico de Comparações")
    print("")
    fig, ax2 = plt.subplots()
    ax2.plot(auxOrdenadoComparacoes, label="Melhor Caso")
    ax2.legend()
    ax2.plot(auxAleatorioComparacoes, label="Aleatorio")
    ax2.legend()
    ax2.plot(auxInversoComparacoes, label="Pior Caso")
    ax2.legend()
    plt.show()

    print("")
    print("1 - c)")
    print("O comportamento assintótico do melhor caso no selection sort é O(n^2)")
    print("O comportamento assintótico do pior caso no selection sort é O(n^2)")
    
main()
```

##  Bubble Sort


```python
def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(0, 100, (n_elementos))
  return A

def tempo(algoritmo, arranjo):
  inicio = time.time()
  comparacoes = algoritmo(arranjo)
  fim = time.time()
  comparacoes = algoritmo(arranjo)
  tempoExecucao = fim - inicio

  return tempoExecucao, comparacoes

def bubbleSort(lis):
    length = len(lis)
    compara = 0
    for i in range(length):
        compara += 1
        for j in range(length - i):
            compara += 1
            a = lis[j]
            if a != lis[-1]:
                compara += 1
                b = lis[j + 1]
                if a > b:
                    lis[j] = b
                    lis[j + 1] = a
    return compara
    
def main():
    auxAleatorio = []
    auxOrdenado = []
    auxInverso = []
    auxAleatorioComparacoes = []
    auxOrdenadoComparacoes = []
    auxInversoComparacoes = []
    tempoAleatorio = 0
    tempoOrdenado = 0
    tempoInverso = 0
    for a in range(5,1000):
      aleatorio = gera_arranjos_aleatorio(a)
      auxAleatorio.append(tempo(bubbleSort, aleatorio)[0])
      auxAleatorioComparacoes.append(tempo(bubbleSort, aleatorio)[1])
      tempoAleatorio += tempo(bubbleSort, aleatorio)[0]
      ordenado = np.arange(a)
      auxOrdenado.append(tempo(bubbleSort, ordenado)[0])
      auxOrdenadoComparacoes.append(tempo(bubbleSort, aleatorio)[1])
      tempoOrdenado += tempo(bubbleSort, aleatorio)[0]
      inverso = np.arange(a, 0, -1)
      auxInverso.append(tempo(bubbleSort, inverso)[0])
      auxInversoComparacoes.append(tempo(bubbleSort, aleatorio)[1])
      tempoInverso += tempo(bubbleSort, aleatorio)[0]

    print("")
    print("Gráfico de Tempo")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxInverso, label="Pior Caso")
    ax.legend()
    ax.plot(auxAleatorio, label="Aleatorio")
    ax.legend()
    ax.plot(auxOrdenado, label="Melhor Caso")
    ax.legend()
    plt.show()

    print("")
    print("Gráfico de Comparações")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxInversoComparacoes, label="Pior Caso")
    ax.legend()
    ax.plot(auxAleatorioComparacoes, label="Aleatorio")
    ax.legend()
    ax.plot(auxOrdenadoComparacoes, label="Melhor Caso")
    ax.legend()
    plt.show()

    print("")
    print("1 - c)")
    print("O comportamento assintótico do melhor caso no bubble sort é O(n)")
    print("O comportamento assintótico do pior caso no bubble sort é O(n^2)")
  
main()
```

## Heap Sort

```python
def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(0, 100, (n_elementos))
  return A

def tempo(algoritmo, arranjo):
  inicio = time.time()
  comparacoes = algoritmo(arranjo)
  fim = time.time()
  tempoExecucao = fim - inicio

  return tempoExecucao, comparacoes

def buildHeap(lista, n):
    compara = 0
    for i in range(n//2 - 1, -1, -1):
        compara += 1
        heapify(lista, n, i)
    comparacoes = heapify(lista, n, i)
    compara += comparacoes
    return compara

def heapify(lista, n, i):
    largest = i  
    left = (2 * i) + 1    
    right = (2 * i) + 2 
    compara = 0
    compara += 1
    if left < n and lista[largest] < lista[left]:
        largest = left
    compara += 1
    if right < n and lista[largest] < lista[right]:
        largest = right
    compara += 1
    if largest != i:
        lista[i], lista[largest] = lista[largest], lista[i] 
        compara += heapify(lista, n, largest) 
    
    return compara

def heapSort(lista):
    n = len(lista)
    buildHeap(lista, n)
    compara = 0
    compara += buildHeap(lista, n)
    for i in range(n-1, 0, -1):
        compara += 1
        lista[i], lista[0] = lista[0], lista[i]
        compara += heapify(lista, i, 0)
    
    return compara

def main():
    auxAleatorio = []
    auxOrdenado = []
    auxInverso = []
    auxAleatorioComparacoes = []
    auxOrdenadoComparacoes = []
    auxInversoComparacoes = []
    tempoAleatorio = 0
    tempoInverso = 0
    tempoOrdenado = 0
    for a in range(5,1000):
      aleatorio = gera_arranjos_aleatorio(a)
      tempoAleatorio = tempo(heapSort, aleatorio)
      auxAleatorio.append(tempo(heapSort, aleatorio)[0])
      auxAleatorioComparacoes.append(tempo(heapSort, aleatorio)[1])
      ordenado = np.arange(a)
      auxOrdenado.append(tempo(heapSort, ordenado)[0])
      auxOrdenadoComparacoes.append(tempo(heapSort, aleatorio)[1])
      tempoOrdenado = tempo(heapSort, ordenado)
      inverso = np.arange(a, 0, -1)
      auxInverso.append(tempo(heapSort, inverso)[0])
      auxInversoComparacoes.append(tempo(heapSort, aleatorio)[1])
      tempoInverso = tempo(heapSort, inverso)

    print("")
    print("Gráfico de Tempo")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxOrdenado, label="Pior Caso")
    ax.legend()
    ax.plot(auxAleatorio, label="Aleatorio")
    ax.legend()
    ax.plot(auxInverso, label="Melhor Caso")
    ax.legend()
    plt.show()

    print("")
    print("Gráfico de Comparações")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxOrdenadoComparacoes, label="Pior Caso")
    ax.legend()
    ax.plot(auxAleatorioComparacoes, label="Aleatorio")
    ax.legend()
    ax.plot(auxInversoComparacoes, label="Melhor Caso")
    ax.legend()
    plt.show()

    print("1 - c)")
    print("O comportamento assintótico do melhor caso no bubble sort é Θ(n log n)")
    print("O comportamento assintótico do pior caso no bubble sort é Θ(n log n)")
    
main()

  
```

## Quick Sort

> pivo = *A[A-comprimento]*

```python

def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(0, 100, (n_elementos))
  return A

def tempo(algoritmo, arranjo):
  inicio = time.time()
  algoritmo(arranjo)
  fim = time.time()
  tempoExecucao = fim - inicio

  return tempoExecucao

def quickSort(array):
    sort(array, 0, len(array) - 1)


def sort(array, left, right):
    if left >= right:
        return array
    pivot = array[(left + right) // 2]
    index = partition(array, left, right, pivot)

    sort(array, left, index - 1)
    sort(array, index, right)


def partition(array, left, right, pivot):
    while left <= right:
        while array[left] < pivot:
            left += 1
        while array[right] > pivot:
            right -= 1

        if left <= right:
            array[left], array[right] = array[right], array[left]
            left += 1
            right -= 1

    return left

def main():
    auxAleatorio = []
    auxOrdenado = []
    auxInverso = []
    tempoAleatorio = 0
    tempoInverso = 0
    tempoOrdenado = 0
    for a in range(5,1000):
      aleatorio = gera_arranjos_aleatorio(a)
      auxAleatorio.append(tempo(quickSort, aleatorio))
      tempoAleatorio += tempo(quickSort, aleatorio)
      ordenado = np.arange(a)
      auxOrdenado.append(tempo(quickSort, ordenado))
      tempoOrdenado += tempo(quickSort, ordenado)
      inverso = np.arange(a, 0, -1)
      auxInverso.append(tempo(quickSort, inverso))
      tempoInverso += tempo(quickSort, inverso)

    print("")
    print("Gráfico de Tempo")
    print("")
    fig, ax = plt.subplots()
    ax.plot(auxInverso, label="Pior Caso")
    ax.legend()
    ax.plot(auxAleatorio, label="Aleatorio")
    ax.legend()
    ax.plot(auxOrdenado, label="Melhor Caso")
    ax.legend()
    plt.show()

    print("1 - c)")
    print("O comportamento assintótico do melhor caso no quick sort é O(n lg n)")
    print("O comportamento assintótico do pior caso no quick sort é O(n^2)")

main()

```

##Quick Sort com pivo aleatório

```python
def tempo(algoritmo, arranjo):
  inicio = time.time()
  algoritmo(arranjo)
  fim = time.time()
  tempoExecucao = fim - inicio

  return tempoExecucao

def quickSortAleatorio(a, inicio=0, fim=None):
    fim = fim if fim is not None else len(a)
    if inicio < fim:
        pp = particao(a, inicio, fim)
        quickSortAleatorio(a, inicio, pp)
        quickSortAleatorio(a, pp + 1, fim)
    return a

def particao(a, inicio, fim):
    rand = randrange(inicio, fim)
    a[rand], a[fim - 1] = a[fim - 1], a[rand]
    pivo = a[fim - 1]
    for i in range(inicio, fim):
        if a[i] > pivo:
            fim += 1
        else:
            fim += 1
            inicio += 1
            a[i], a[inicio - 1] = a[inicio - 1], a[i]
    return inicio - 1

def main():
    auxAleatorioQuickSortAleatorio = []
    auxOrdenadoQuickSortAleatorio = []
    auxInversoQuickSortAleatorio = []
    auxAleatorioQuickSort = []
    auxOrdenadoQuickSort= []
    auxInversoQuickSort = []
    tempoAleatorioQuickSortAleatorio = 0
    tempoInversoQuickSortAleatorio = 0
    tempoOrdenadoQuickSortAleatorio = 0
    tempoOrdenadoQuickSort = 0
    tempoAleatorioQuickSort = 0
    tempoInversoQuickSort = 0
    tempoOrdenadoQuickSort = 0
    aleatorio = gera_arranjos_aleatorio(100)
    auxAleatorioQuickSortAleatorio.append(tempo(quickSortAleatorio, aleatorio))
    tempoAleatorioQuickSortAleatorio += tempo(quickSortAleatorio, aleatorio)
    auxAleatorioQuickSort.append(tempo(quickSort, aleatorio))
    tempoAleatorioQuickSort += tempo(quickSort, aleatorio)
    ordenado = np.arange(100)
    auxOrdenadoQuickSortAleatorio.append(tempo(quickSortAleatorio, ordenado))
    tempoOrdenadoQuickSortAleatorio += tempo(quickSortAleatorio, ordenado)
    auxOrdenadoQuickSort.append(tempo(quickSort, ordenado))
    tempoOrdenadoQuickSort += tempo(quickSort, ordenado)
    inverso = np.arange(100, 0, -1)
    auxInversoQuickSortAleatorio.append(tempo(quickSortAleatorio, inverso))
    tempoInversoQuickSortAleatorio += tempo(quickSortAleatorio, inverso)
    auxOrdenadoQuickSort.append(tempo(quickSort, inverso))
    tempoInversoQuickSort += tempo(quickSort, inverso)

    print("")
    print("Comparação")
    print("")
    print("Quick sort com pivô aleatório")
    print("")
    print("O tempo no melhor caso foi de {:.5f}" .format(tempoInversoQuickSortAleatorio))
    print("")
    print("O tempo no pior caso foi de {:.5f}" .format(tempoOrdenadoQuickSortAleatorio))
    print("")
    print("O tempo no caso aleatório foi de {:.5f}" .format(tempoAleatorioQuickSortAleatorio))
    print("")

    print("")
    print("Quick sort")
    print("")
    print("O tempo no melhor caso foi de {:.5f}" .format(tempoOrdenadoQuickSort))
    print("")
    print("O tempo no pior caso foi de {:.5f}" .format(tempoInversoQuickSort))
    print("")
    print("O tempo no caso aleatório foi de {:.5f}" .format(tempoAleatorioQuickSort))
    print("")

main()    
```
